```python
import pandas as pd
import numpy as np
import matplotlib as plt 
import matplotlib.pyplot as plt
import statsmodels
```

##### Importing data

```python
df = pd.read_csv('JetRail_Train_SU63ISt.csv')
df.head()
```

```python
df.tail()
```

2 years of hourly data from Aug 2012 to Sep 2014

##### Aggregating the dataset up to a daily level

```python
#possible_formats = ['%d-%m-%Y %H:%M', '%d/%m/%Y %H:%M', '%d/%m/%Y %I:%M %p']
```

```python
#for format in possible_formats:
    try:
        train['Timestamp'] = pd.to_datetime(train['Datetime'], format) #Try to get the date
        break #If correct, don't try any other formats 
    except ValueError:
        pass #If incorrect format, keep trying other formats
```

```python
df['Timestamp'] = pd.to_datetime(df.Datetime, format = '%d-%m-%Y %H:%M')
```

```python
df.index = df.Timestamp
```

```python
df = df.resample('D').mean()
```

```python
df.tail()
```

```python
df.head()
```

Subsetting the dataset from Aug 2012 - Jul 2014

```python
dateRange = (df.index > '2012-08-24') & (df.index <= '2014-08-01')
```

```python
train = df.iloc[dateRange]
```

```python
test = df.iloc[~dateRange]
```

```python
print(train.head(), train.tail(), test.head(), test.tail())
```

```python
train.Count.plot(figsize=(15,8), title= 'Daily Ridership', fontsize=14)
test.Count.plot(figsize=(15,8), title= 'Daily Ridership', fontsize=14)
plt.show()
```

#### Naive Model

```python
dd = np.asarray(df['Count'])
y_hat_df = test.copy()
y_hat_df['naive'] = dd[len(dd)-1]
```

```python
plt.figure(figsize = (12,8))
plt.plot(train.index, train['Count'], label = 'train')
plt.plot(test.index, test['Count'], label = 'test')
plt.plot(y_hat_df.index, y_hat_df['naive'], label = 'Naive forecast')
plt.legend(loc='best')
plt.title('Naive Forecast')
plt.show()
```

##### RMSE of Naive Model

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(test['Count'], y_hat_df['naive']))
print(rmse)
```

#### Simple average 

```python
y_hat_df['avg_forecast'] = train['Count'].mean()
plt.figure(figsize= (12,8))
plt.plot(train.index, train['Count'], label = 'train')
plt.plot(test.index, test['Count'], label = 'test')
plt.plot(y_hat_df.index, y_hat_df['avg_forecast'], label = 'Simple average forecast')
plt.legend(loc = 'best')
plt.title('Simple average forecast')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test['Count'], y_hat_df['avg_forecast']))
print(rmse)
```

#### Moving average 

```python
y_hat_df['moving_avg_forecast'] = train['Count'].rolling(60).mean().iloc[-1] #Avg of last 60 pts of train is the forecast for test
```

```python
plt.figure(figsize=(16,8))
plt.plot(train.Count, label='train')
plt.plot(y_hat_df.Count, label = 'test')
plt.plot(y_hat_df.moving_avg_forecast, label = 'Moving avg forecast')
plt.legend(loc = 'best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(y_hat_df['Count'], y_hat_df['moving_avg_forecast']))
print(rmse)
```

#### Simple exponential smoothing

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
fit2 = SimpleExpSmoothing(np.asarray(train['Count'])).fit(smoothing_level=0.6,optimized=False)
y_hat_df['SES'] = fit2.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_df['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test.Count, y_hat_df.SES))
print(rmse)
```

#### Holt's linear trend method

```python
import statsmodels.api as sm
sm.tsa.seasonal_decompose(train['Count']).plot()
result = sm.tsa.stattools.adfuller(train['Count'])
plt.show()
```

We can also generate a multiplicative forecast equation by multiplying trend and level instead of adding it. When the trend increases or decreases linearly, additive equation is used whereas when the trend increases of decreases exponentially, multiplicative equation is used.Practice shows that multiplicative is a more stable predictor, the additive method however is simpler to understand.
Ref: http://www.forsoc.net/2014/11/11/can-you-identify-additive-and-multiplicative-seasonality/

```python
fit1 = Holt(np.asarray(train['Count'])).fit(smoothing_level = 0.3,smoothing_slope = 0.1)
y_hat_df['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_df['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test.Count, y_hat_df.Holt_linear))
print(rmse)
```

#### Holt Winters Method

```python
fit1 = ExponentialSmoothing(np.asarray(train['Count']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_df['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_df['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test.Count, y_hat_df.Holt_Winter))
print(rmse)
```
